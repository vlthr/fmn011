```python
def bisection (func, a, b, tol=1e-12, max_iterations=50):
    iterations = 0
    left = a
    right = b
    
    while ((right - left)/2 > tol and iterations < max_iterations):
        iterations+=1
        c = (left + right)/2
        if (func(c) * func(left) > 0):
            left = c
        elif (func(c) * func(right) > 0):
            right = c
    
    c = (left + right)/2
    poss_err = (right - left)/2
    return c
```

```python
def fixed_point (func ,start_p, tol=1e-12, max_itr=1000):
    iterations = 0
    p = start_p
    while (np.abs(p - func(p)) > tol and iterations < max_itr):
        iterations+=1
        p = func(p)
        
    print(iterations)
    return p
```
